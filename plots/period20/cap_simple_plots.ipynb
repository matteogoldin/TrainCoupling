```python
import matplotlib.pyplot as plt 
import numpy as np

#Capacità semplice, timeLimit=300, runsNumber=10000 
capArray_5h = [7.5482000000000005, 10.436000000000002, 13.229700000000001, 15.906, 15.0, 16.5484, 14.0, 15.984999999999998, 10.180900000000003, 10.0]
N = list(range(1,len(capArray_5h)+1))

plt.plot(N, capArray_5h, "-o", color ='orange')
  
# naming of x-axis and y-axis
plt.xlabel('Num treni accoppiati')
plt.xticks(np.arange(1, len(N)+1, 1))
plt.ylabel('Capacità')
  
# naming the title of the plot
plt.title("Stima della capacità del tratto con train coupling in 5 ore")

plt.grid()

plt.savefig("capacities_5h.png", format="png", dpi=160)

plt.show()
```
