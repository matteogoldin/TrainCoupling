```python
import matplotlib.pyplot as plt 
import numpy as np

#Capacità semplice, timeLimit=300, runsNumber=10000 
capArray_simple_5h = [7.4997, 10.364399999999998, 12.551300000000001, 13.941699999999999, 12.736800000000002, 14.134800000000004, 10.476199999999999, 11.987699999999998, 9.457500000000005, 10.0]
capArray_advanced_5h = [7.5482000000000005, 10.436000000000002, 13.229700000000001, 15.906, 15.0, 16.5484, 14.0, 15.984999999999998, 10.180900000000003, 10.0]
N = list(range(1,len(capArray_advanced_5h)+1))

plt.plot(N, capArray_advanced_5h, "-o", color ='orange', label ='avanzata')
plt.plot(N, capArray_simple_5h, "-o", color ='blue', label ='semplice')
  
# naming of x-axis and y-axis
plt.xlabel('Num treni accoppiati')
plt.xticks(np.arange(1, len(N)+1, 1))
plt.ylabel('Capacità')
  
# naming the title of the plot
plt.title("Stima della capacità del tratto con train coupling in 5 ore")
  
plt.legend()
plt.grid()

plt.savefig("capacities_5h.png", format="png", dpi=160)

plt.show()
```
